```python
import pandas as pd
```

```python
RESULT_PATHs = [
    "../grid_searches/deepsurv/miRNA/clinical_miRNA_normalized_log/cv_results_Net_3layers.csv",
    "../grid_searches/deepsurv/miRNA/clinical_miRNA_normalized_log/cv_results_Net_5layers.csv",
    "../grid_searches/deepsurv/miRNA/clinical_miRNA_normalized_quant/cv_results_Net_3layers.csv",
    "../grid_searches/deepsurv/miRNA/clinical_miRNA_normalized_quant/cv_results_Net_5layers.csv",
    "../grid_searches/deepsurv/mRNA/clinical_mRNA_normalized_log/cv_results_Net_3layers.csv",
    "../grid_searches/deepsurv/mRNA/clinical_mRNA_normalized_log/cv_results_Net_5layers.csv",
    "../grid_searches/deepsurv/mRNA/clinical_mRNA_normalized_tpm_log/cv_results_Net_3layers.csv",
    "../grid_searches/deepsurv/mRNA/clinical_mRNA_normalized_tpm_log/cv_results_Net_5layers.csv",    
]
for RESULT_PATH in RESULT_PATHs:
    df = pd.read_csv(RESULT_PATH)
    
    # calculate mean and all information distribution for c-index
    mean_cindex = df['c_index'].mean()
    std_cindex = df['c_index'].std()
    min_cindex = df['c_index'].min()
    max_cindex = df['c_index'].max()
    p25_cindex = df['c_index'].quantile(0.25)
    p50_cindex = df['c_index'].quantile(0.50)
    p75_cindex = df['c_index'].quantile(0.75)
    
    print(f"CSV path: {RESULT_PATH}\n")
    print(f"Mean c-index: {mean_cindex:.4f}")
    print(f"Std c-index: {std_cindex:.4f}")
    print(f"Min c-index: {min_cindex:.4f}")
    print(f"Max c-index: {max_cindex:.4f}")
    print(f"25th Percentile c-index: {p25_cindex:.4f}")
    print(f"50th Percentile c-index: {p50_cindex:.4f}")
    print(f"75th Percentile c-index: {p75_cindex:.4f}")
    print("\n=========================================================\n")
```
