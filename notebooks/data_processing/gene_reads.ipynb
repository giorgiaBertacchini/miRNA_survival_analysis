```python
import pandas as pd
import os
import ast

CLINICAL_miRNA_PATH = '../../datasets/preprocessed/clinical_miRNA(RC_RPM).csv'
```

```python
clinical_miRNA_df = pd.read_csv(CLINICAL_miRNA_PATH)
print(clinical_miRNA_df.head(2))
```

```python
clinical_miRNA_df["miRNA_ID"] = clinical_miRNA_df["miRNA_ID"].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
clinical_miRNA_df["reads_per_million_miRNA_mapped"] = clinical_miRNA_df["reads_per_million_miRNA_mapped"].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x
)

# Create Dict {sample_name: {miRNA: value}}
data_dict = {}
for i, row in clinical_miRNA_df.iterrows():
    sample_name = f"Sample_{i}"
    mirna_ids = row["miRNA_ID"]
    mirna_values = row["reads_per_million_miRNA_mapped"]
    
    data_dict[sample_name] = dict(zip(mirna_ids, mirna_values))

expression_df = pd.DataFrame.from_dict(data_dict, orient="index")

print(expression_df.head())
```

```python
expression_df.to_csv(os.path.join('../../datasets/preprocessed/genes_reads(RPM).csv'))
```
